5. **Burials per Decade**  
    a. Create a "Decade" column to show the decade in which a burial occurred.    
    b. Examine the burials per decade. Look at the total number of burials and the proportion (aka percentage) of male burials and female burials.  
    c. Create a clustered bar chart to show how male and female burials have changed over time.

```python
city_clean_final = city_clean.dropna(subset=['Burial Year'])
```

```python
city_clean_final
```

```python
city_clean_final['Year'] = pd.to_datetime(city_clean_final['Burial Year'].astype(int), format='%Y').dt.year
```

```python
city_clean_final
```

a. Create a "Decade" column to show the decade in which a burial occurred.    

```python
city_clean_final['Decade'] = city_clean_final['Year'] - city_clean_final['Year']%10
```

```python
city_clean_final
```

```python
city_clean_final_check = pd.crosstab(city_clean_final['Decade'], city_clean_final['Year'])
```

```python
city_clean_final_check
```

```python
city_clean_final
```

```python
b. Examine the burials per decade. Look at the total number of burials and the proportion (aka percentage) of male burials and female burials.  
```

```python
city_clean_male_female = pd.crosstab(city_clean_final['Decade'], city_clean_final['Sex'], normalize='index')
```

```python
city_clean_male_female
```

```python
city_clean_malefemale = city_clean_final[(city_clean_final['Sex'] == 'M') | (city_clean_final['Sex'] == 'F')]
```

```python
city_clean_malefemale
```

```python
city_clean_malefemale_chart = city_clean_malefemale.groupby('Decade')['Sex'].value_counts()
```

```python
city_clean_malefemale_chart
```

c. Create a clustered bar chart to show how male and female burials have changed over time.

```python
city_clean_malefemale_chart.unstack().plot(kind='bar', stacked=False)
```

```python
city_clean_malefemale_chart_normalized = city_clean_malefemale.groupby('Decade')['Sex'].value_counts(normalize = True)
```

```python
city_clean_malefemale_chart_normalized.unstack().plot(kind='bar', stacked=False)
```
